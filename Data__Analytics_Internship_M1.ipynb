```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print("Libraries successfully imported!")
```

```python
# Hum Titanic ka data load kar rahe hain
df = sns.load_dataset('titanic')

# Ye dekhne ke liye ke data kaisa dikhta hai
df.head()
```

```python
# Isse rows aur columns ki tadad pata chalegi
print("Total Rows and Columns:", df.shape)
```

```python
# Ye check karega ke kis column mein kitna data missing hai
print(df.isnull().sum())
```

```python
# Age ki missing values ko Average (mean) age se fill karte hain
df['age'] = df['age'].fillna(df['age'].mean())

# Deck mein bohat zyada data missing hai, toh hum us column ko hi khatam (drop) kar dete hain
df.drop(columns=['deck'], inplace=True)

# Ab dobara check karte hain ke cleaning hui ya nahi
print(df.isnull().sum())
```

```python
# Gender ke hisab se survival rate check karte hain
print(df.groupby('sex')['survived'].mean())
```

```python
# Survived vs Not Survived ka graph
plt.figure(figsize=(8,5))
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival Count by Gender')
plt.show()

# Age distribution dekhne ke liye histogram
plt.figure(figsize=(8,5))
sns.histplot(df['age'], kde=True, color='purple')
plt.title('Age Distribution of Passengers')
plt.show()
```
