```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


print("Libraries successfully imported!")
```

```python
# we load titanic data
df = sns.load_dataset('titanic')

# we want to se how data looks like
df.head()
```

```python

print("Total Rows and Columns:", df.shape)
```

```python
# we check how much data missing in columns
print(df.isnull().sum())
```

```python
# we check how much data missing in columns
print(df.isnull().sum())
```

```python

df['age'] = df['age'].fillna(df['age'].mean())


df.drop(columns=['deck'], inplace=True)


print(df.isnull().sum())
```

```python
# survival rate
print(df.groupby('sex')['survived'].mean())
```

TASK 2 SUMMARY
* Data Cleaning: Loaded the Titanic dataset and handled missing values. 'Age' was filled with the mean value, and the 'Deck' column was dropped due to excessive missing data.

* Data Structure: Used info() and shape to understand the dimensions and data types of the dataset.

* Key Insight: Analysis revealed that passenger gender and ticket class were the most significant factors influencing survival.   
  








# Task 3: Data Visualization & InsightsWhatsApp Image 2026-04-15 at 2.54.12 AM.jpeg

```python
# Survived vs Not Survived graph
plt.figure(figsize=(4,3))
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

```python
# Age distribution  histogram
plt.figure(figsize=(5,3))
sns.histplot(df['age'], kde=True, color='pink')
plt.title('Age Distribution of Passengers')
plt.show()
```

A. Countplot (Survived vs Not Survived) :

Observation: This graph shows that the number of people who did not survive is higher than those who survived. It gives us a clear picture of the overall tragedy.

B. Histogram (Age Distribution) :

Observation: Most of the passengers on the ship were young adults, specifically between the ages of 20 and 40. There were also some children and elderly people present.

```python
# Pclass (Passenger Class) ke hisab se survival rate
plt.figure(figsize=(5,3))
sns.barplot(x='pclass', y='survived', data=df)
plt.title('Survival Rate by Ticket Class')
plt.show()
```

Barplot (Survival by Class) :

* Observation: Passengers in the 1st Class had the highest chance of survival, while those in the 3rd Class had the lowest. This shows that socio-economic status played a role in survival.

```python

# Pie Chart for Overall Survival
plt.figure(figsize=(3,3))
df['survived'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], labels=['Not Survived', 'Survived'], explode=(0.05,0))
plt.title('Overall Survival Percentage')
plt.ylabel('')
plt.show()
```

Pie Chart:
* Observation: Out of all passengers, only 38.4% survived, while 61.6% lost their lives.
